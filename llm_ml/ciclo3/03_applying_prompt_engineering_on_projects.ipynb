# 3 Aplicación de mejores prácticas de prompt engineering en proyectos digitales

## 1. Objetivo

En seguimiento al reporte previo relativo a principios y mejores prácticas de prompt engineering, el equipo de SmartMedio realizó un tercer reporte interno de investigación con objetivo de explorar cómo pueden aplicarse los anteriores en proyectos digitales de modo que se pueda optimizar el diseño de prompt para obtener mejores resultados a las preguntas de interés y que será un documento de apoyo para el proceso de prompt engineering.

# 2 Aplicaciones en proyectos digitales

## 2.1  Relacionados con generación de texto

Los modelos generativos de lenguaje son muy versatiles al tener como base de entrenamiento a grandes conjuntos de datos y contar con la capacidad de generar respuestas coherentes y relevantes. Estas características permiten incorporar su uso en proyectos digitales en múltiples formas, a continuación se listan algunas de ellas:

* **Generación de contenido:** Típicamente los proyectos digitales necesitan desarrollar material (e.g. slogans, textos creativos, artículos y descripciones de productos) que será mostrado sitio web, redes sociales, anuncios o campañas publicitarias, el cual se pueden consolidar desde un modelo LLM.
* **Correción de texto:** la revisión gramática, redacción y reescritura de textos alrededor de un proyecto puede ser solicitada a una de dichas herramientas acelerando los procesos de control de calidad,

* **Generación de ideas o instrucciones:** para el desarrollo de un proyecto de cero o bien en momentos de bloqueo creativo, se pueden pedir ideas relevantes para una tarea y los pasos necesarios para llevarlas a cabo,

* **Educación/tutoría:** tales pueden fungir  como tutor virtual, brindando explicaciones y respuestas a preguntas de los usuarios sobre diversos temas. Este punto se puede ampliar a que las herramientas sea compañeros de conversación que nos expliquen temas o nos retroalimente en una conversación sobre un tema, por ejemplo desarrollando código o aprendiendo un idioma.

* **Creación de código:** dichas herramientas pueden crear código de muchos lenguajes de programación (como Python, R, SQL, Scala, Javascript) y son útiles para generer versiones rápidas de sitios web, aplicaciones y otros tipos de software.


## 2.2  Relacionados con generación de imágenes

Una de las características más atractivas de los modelos generativos es que puede crear imágenes muy detallas a partir de los elementos que se describen en el prompt, sin embargo al tratarse de elementos estéticos suele existir una componente subjetiva para juzgar los resultados. Considerando tales puntos, estas ideas se pueden incorporar en proyectos digitales en diferentes vertientes:

* **Generación de prototipos y diseños conceptuales:** En proyectos de diseño, se puede empelar tales herramientas para generar rápidamente prototipos y diseños conceptuales para evaluar diferentes diseños de estilo rápidamente.

* **Generación de contenido multimedia:** Se puede emplear el modelo para crear contenido como ilustraciones para libros, portadas de álbumes, gráficos para redes sociales o imágenes para blogs.

* **Creación de personajes y avatares:** Si se está desarrollando un creatividades para un personaje, una plataforma de realidad virtual o un entorno en línea, un modelo generativo puede ser utilizado para generar personajes y avatares únicos. Esto permite a los usuarios personalizar sus representaciones virtuales de una manera más diversa.

* **Diseño de productos personalizados:** Si se cuenta con un negocio de productos susceptibles de personalización (ropa, accesorios, muebles u otros) puedes utilizar un modelo para permitir a los clientes crear sus propios diseños y potenciar el interés en la mercancía.

* **Generación de arte y diseño:** tales herramientas se pueden utilizar para crear obras de arte y diseños. Son ideales pues permiten experimentar para varias estilos, colores y formas para obtener resultados interesantes que luego pueden ser utilizados en productos, campañas publicitarias o incluso en el diseño de páginas web.
