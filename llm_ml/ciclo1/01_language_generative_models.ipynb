# 2. Elementos de modelos generativos y Large Language Models (LLM)

## 2.1 Modelos Generativos

La Inteligencia Artificial es una rama de conocimiento con gran augente en las últimas décadas y que engloba particularmente a un área llamada **Procesamiento de Lenguaje Natural** (NLP, por sus siglas en inglés), que se ocupa de cómo las máquinas pueden interactuar con el lenguaje humano en diferentes modalidades, como texto, audio, imágenes y otros medios, de manera similar a las personas.

En los últimos años, el NLP ha sido aplicado a muchas soluciones tecnológicas, destacando en áreas como la traducción automática, el reconocimiento de voz y la clasificación de texto. En particular, se ha dado relevancia al subcampo de generación de lenguaje, que busca desarrollar algoritmos o soluciones para crear texto, imágenes, videos y otras expresiones lingüísticas a partir de datos de entrada en texto.

Tales desarrollos son posibles gracias a una combinación de técnicas de aprendizaje de máquina, que permiten a un modelo computacional **aprender** a partir de una cantidad grande de ejemplos la estructura de la información que se le presentan y con dicha abstracción de conocimiento generar una respuesta relevante a una indicación o pregunta que introduzcan los usuarios mediante texto. 

**Figura 1: Esquema de un modelo generativo**

![title](./images/generative_model.png)

Entre varios tipos de modelos generativos:

* **Texto-a-Texto:** Reciben instrucciones de entrada en texto (e.g. preguntas o instrucciones) y generan un texto como respuesta. La idea de su diseño es que puedan capturar las posibles relaciones entre los textos de entrada y el texto generado como salida.
* **Texto-a-Imagen:** Se especializan en generar una imagen a partir de una descripción en texto
* **Texto-a-Video:** Son modelos que generan respuestas en video a las descripciones en texto que se les proveen.

## 2.2 ¿Cómo funcionan los modelos generativos?

A nivel teórico, podemos pensar que los modelos generativos se especializan en aprender la información subyacentes de un conjunto de datos que se les proveen como ejemplo y luego usar ese conocimiento para generar nuevos datos relacionados con las preguntas que reciben como entrada.

Por ello, mucha de la investigación reciente en ésta línea se basa en modelos de aprendizaje profundo que emplean redes neuronales, que no son más que modelos computacionales inspirados en el cerebro humano y están formadas por capas de neuronas; cada neurona recibe entradas de otras neuronas y produce una salida. Las salidas de todas las neuronas de una capa se utilizan como entradas para las neuronas de la siguiente capa y el resultado final de la arquitectura su usa para realizar una acción relacionada al problema que queremos resolver. Idealmente, el conjunto de datos de entrenamiento sirve para calibrar interativamente un conjunto de datos numéricos ligado a los modelos, que se denomínan parámetros o pesos, hasta alcanzar valores óptimos que permiten hacer predicciones precisas.

**Figura 2: Esquema de una red neural usada en aprendizaje profundo**

![title](./images/dl_layer.png)

La gran ventaja de los modelos de aprendizaje profundo reside en que se puede adaptar el diseño de su arquitectura para capturar patrones complejos de información, como la que se encuentra presente en el nuestro lenguaje, pensemos por ejemplo en la posición de las palabras, el significado y la similaridad entre dos textos aún siendo distintos. 

La explicación anterior, por supuesto, nos hace notar los principios de los que depende un modelo generativo:

1) *Datos de entrenamiento:* son conjuntos de datos, generalmente muy grandes, que se necesitan como ejemplos para entrenar el modelo. Las respuestas generadas por el modelo dependen de los ejemplos provistos, por lo que se pueden entrenar modelos para tareas de campos específicos.
2) *Arquitectura de redes neuronales:* Se refiere al tipo de las redes neuronales profundas, que permiten comprender y reconocer las relaciones y conexiones entre las palabras y los conceptos reflejados en los datos de entrenamiento del modelo. Éste tema ha sido objeto amplio de investigación durante varias décadas, siendo los [*transformers*](https://ai.googleblog.com/2017/08/transformer-novel-neural-network.html) piezas fundamentales para el auge visto recientemente en modelos generativos de texto, que gracias a su mecanismo de atención permiten capturar estructuras presentes en el texto como el orden y similiridad de significados. Cabe destacar que generalmente se calibran diseños de redes neuronales que pueden contar con hasta billones de parámetros.
3) *Mecanismo de evaluación del contenido generado*: esencialmente se trata de una estrategia para entender si el contenido generado por el modelo es bueno o no, e incorporar un cierto grado de retroalimentanción como componente del modelo. Aunque muchos enfoques se han intentado, una técnica que ha demostrado buenos resultados es el [RLHF (*Reinforcement Learning with Human Feedback*)](https://huggingface.co/blog/rlhf) que en resumen introduce un marco de aprendizaje por refuerzo, que permite premiar o penalizar contenido generado por el modelo y que no es coherente con la tarea que se quiere resolver, para después ajustar nuevamente los parámetros incorporando dicho conocimiento.

Adicionalmente, podemos añadir que en la práctica obtener un buen resultado depende además de:

4) *Diseño del texto de entrada:* dado que este tipo de modelo requiere patrones de entrada a nivel de texto, es natural pensar que un diseño adecuado del texto que se le introduce, va a mejorar la calidad del contenido generado. En el campo de trabajo, las prácticas y técnicas para optimizar el diseño de las instrucciones que le damos a modelos generativos se denomina **prompt engineering**.

## 2.3 ¿Qué es un Large Language Model?

Hablando nuevamente de modelos generativos texto-a-texto, es importante mencionar el éxito de los modelos denominados *Large Language Model*. En esencia, se trata de modelos que abordan la generación de contenido a partir de texto de entrada o de un conjunto de instrucciones dadas por un usuario, como un problema de predicción de sequencias de texto; es decir, tales herramientas intenta predecir cual es el texto que sea más acorde a una colección de texto ordenado.

Para ilustralo, pensemos que queremos entrenar un modelo que sea capaz de responder cual es la siguiente palabra de la oración "Añade a la olla la". Proveyendo suficientes ejemplos, quizá incluyendo textos que tengan que ver con temas de temas de la vida cotidiana, podriamos pensar en respuestas como a) llanta, b) dinosaurio, c) Saturno, d) cebolla. Intutivamente sabemos que con los incisos a) y d) podríamos tener una oración coherente, pero el primer inciso redundaria en una oración extraña (*Añade a la olla la llanta*), mientras que el resto de incisos no parece tener relación alguna con un texto con significado aceptable en el idioma. En este caso un modelo de tipo LLM, calcularía un valor de probabilidad de cual es la palabra que sigue a la oración de entrada para generar el texto correspondiente. 

Ésta idea sencilla que es la base con la que funciona dicha arquitectura de modelos, es central para consolidar buenas prácticas de *prompt engineering*, dado que intutivamente nos dice que si queremos obtener resultados adecuados al darle instrucciones a un modelo generativo, necesitamos ser cuidados en los patrones de texto que introducimos, así el modelo generará texto resultado a partir de predicciones basadas en el contexto y la estructura que le hemos provisto de inicio.

Por otro lado, vale la pena destacar que estos modelos se asocian a la palabra inglesa *Large* por dos motivos:

1) Generalmente necesitan un cantidad de datos ejemplo de volumen enorme, como datos de foros de internet o post de redes sociales y
2) Cuentan con una cantidad enorme de parámetros a ser calibrados (los modelos actuales rondan el orden de billones de parámetros). Por su puesto, un entrenamiento de modelos requiere de amplio poder computacional.

Cabe destacar que en la actualidad, los modelos generativos y los LLM son un campo activo de investigación. De hecho, trabajos recientes de investigación también se han enfocado en desarrollar modelos LLM a través de los cuales se puedan sostener conversaciones coherentes e informativas entre personas y máquinas.

# 4.  Elementos básicos sobre prompts y sus resultados

En todas las plataformas anteriores poseen interfaces de ChatBot donde el usuario introduce texto escrito y se permite obtener la respuesta inmediata de las instrucciones señaladas. Al texto de entrada suele denominársele *prompt* y, como se han mencionado anteriormente la estructura del texto que se introduce para realizar un tarea afecta a los resultados que se pueden obtener con este tipo de herramientas.

![title](./images/prompt_diagram.png)

En este sentido, no se debe perder de vista que, en términos generales, los modelos LLM tratan de resolver un problema de predicción de las palabras que continuan a una secuencia de texto, con lo cual, debemos ser cuidadosos con el contenido de esa cadena de entrada para obtener un resultado coherente e informativa para la consulta de interés.

## 4.2 Errores comunes al trabaja con modelos generativos y LLM

Un punto importante a considerar tanto en el diseño de prompts como en la interpretación de los resultados es que los modelos pueden tener errores y sesgos provemientes de multiples fuentes:

a) **Datos entrenamiento:** los modelos son entrenados generalmente con información presente en obras de texto, foros de Internet (como Reddit) y redes sociales de los cuales no exista un arbitraje, con lo cual las respuestas construidas a partir de dicha información pueden ser imprecisas, erroneas, no corresponder con la realidad o bien reflejar puntos de vistas tendenciosos.

b) **Temporalidad de los datos:** las arquitecturas actuales de modelos generativos se entrenan previamente a ser usan en aplicaciones de software, es decir, no necesariamente toman fuentes en tiempo real por lo que existe un desfase temporal en la información que dichas herramientas conocen. Sus respuestas están atadas a fuentes de datos en el pasado, por lo tanto pueden responder información incompleta o simplemente desconocer la actualidad de los sucesos (por ejemplo: un modelo generativo entrenado en 2021, no tiene información de quién fue el equipo campeón del mundo en 2022). Una forma de lidiar con este fenómeno es darle al modelo los datos particulares que necesitamos que evalue, a manera de contexto o ejemplos.

c) **Entrenados para un propósito:** cómo se ha mencionado, los modelos generativos se entrenan para resolver problemás específicos (por ejemplo, conversar con una persona de forma coherente), por lo tanto no tendrán buenos resultados con tareas ajenas para los que fueron entrenados. En este sentido, los modelos LLM suelen tener desempeños desafortunados en tareas como resolver problemas matemáticos o responder a cuestiones que no están contenidas en los datos con los que se entrenaron.

d) **Alucinaciones:** Un fenómeno detectado en los modelos generativos es que en ocasiones generan de contenido que puede parecer real o verdadero, pero en realidad no está basado en información o conocimientos verificables. En otras palabras, los modelos puede producir respuestas o información que no están respaldadas por hechos o que no existen en la realidad. Aunque las razones detrás de estos hechos aun son tema de investigación, se sabe que las alucinaciones provienen de ser entrenados con pocos datos, la arquitectura del modelo puede no capturar sufientemente bien las relaciones semánticas en el lenguaje o puede faltar etapas de ajuste en su diseño.

e) **Riesgos de seguridad:** Típicamente usamos los modelos generativos a través de una plataforma pública o privada, que tiene acceso a los prompts y la información que le compartimos a las herramientas a través de las preguntas. En este sentido, compartir información personal o sensible a un ChatBot puede exponer a las personas o empresas al exponer información que no necesariamente debería estar en poder de un tercero. Adicionalmente, considerando los errores, sesgos y alucionaciones de los modelos generativos, la interpretación de los resultados de un modelo debería siempre verificando que las soluciones sean verificables y que no impliquen riesgos a personas y organizaciones.

**Otras limitaciones**

* **Limitaciones de tokens:** Usualmente los plataformas proveedoras de ChatBots basados en LLM ponen restricciones a la cantidad de palabras que podemos comunicar como datos de entrada y también al tamaño producido por un modelos. Con ello, las instrucciones que comunicamos y la información adicional que se le da a un modelo debe ser concisa.

* **Aleatoreidad en la respuestas generadas por modelo generativa de lenguaje:** En general, los modelos generativos y especialmente los LLM emplean parámetros de "temperatura" que controlan la función de probabilidad de las salidas del modelo, es decir, limitan que tan aleatoria es la respuesta generada como la salida. Podemos pensar que una temperatura más alta dará como resultado una salida más creativa y menos predecible, mientras que una temperatura más baja dará como resultado una salida más conservadora y más similar a la entrada. En este sentido, ante un mismo prompt podemos obtener respuestas distintas.
